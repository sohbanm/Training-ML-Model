# Task 5, preprocessing pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_cols = X.select_dtypes(include='number').columns.to_list()
cat_cols = X.select_dtypes(exclude='number').columns.to_list()

# Create pipelines for numeric and categorical columns
num_pipeline = make_pipeline(SimpleImputer(strategy='mean'), StandardScaler())
cat_pipeline = make_pipeline(SimpleImputer(strategy='most_frequent'), OneHotEncoder(sparse_output=False))

# Use ColumnTransformer to set the estimators and transformations

preprocessing = ColumnTransformer([('num', num_pipeline, num_cols),
                                   ('cat', cat_pipeline, cat_cols)],
                                    remainder='passthrough'
                                 )
preprocessing
```

```python
# Apply the preprocessing pipeline on the dataset

X_prepared = preprocessing.fit_transform(X)

# Scikit-learn strips the column headers, so just add them back on afterward.
feature_names=preprocessing.get_feature_names_out()
X_prepared = pd.DataFrame(data=X_prepared, columns=feature_names)

X_prepared
```

# Task 6 target value cleaning

```python
y.value_counts()
```

# Task 7 remove period

```python
y = y.replace('<=50K.', '<=50K', regex=False)
y = y.replace('>50K.', '>50K', regex=False)
y.value_counts()
```

# Task 8 split data 80/20

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_prepared, y, test_size=0.2, random_state=42)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

# Task 9 train model

```python
from sklearn.svm import SVC

# Instantiate the SVM model with the specified parameters
model_svm = SVC(kernel='poly', gamma=1, C=0.1)

# Train the model on the training set
model_svm.fit(X_train.iloc[:10000], y_train.iloc[:10000].values.ravel())
```

```python
from sklearn.metrics import classification_report

# Make predictions on the test set
y_predict = model_svm.predict(X_test)

# Generate and print the classification report
report = classification_report(y_test, y_predict)
print(report)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_predict)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model_svm.classes_)
disp.plot(cmap='Blues')  # You can customize the colormap if needed
plt.show()
```

# Task 10 GridSearchCV

```python
X_train, X_validation_test, y_train, y_validation_test = train_test_split(X_prepared, y, test_size=0.4, random_state=42)
X_validation, X_test, y_validation, y_test = train_test_split(X_validation_test, y_validation_test, test_size=0.5, random_state=42)

print(X_train.shape, y_train.shape, X_validation.shape, y_validation.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

svm_parameters = {'kernel': ['rbf'],
                  'C': [0.01, 0.1, 1, 10],
                  'gamma': [0.01, 1, 10]
                  }

svm = SVC()
svm_gs = GridSearchCV(estimator=svm, param_grid=svm_parameters, cv=5)
svm_gs.fit(X_train.iloc[:10000],  y_train.iloc[:10000].values.ravel())

# Get the best estimator from the grid search
svm_winner = svm_gs.best_estimator_

# Evaluate the model on the validation set
validation_accuracy = svm_winner.score(X_validation, y_validation)
print(f"Validation Accuracy with Best Parameters: {validation_accuracy:.4f}")

# Check the svm winner parameters
print("SVM Winner Parameters:")
print(svm_winner.get_params())
```
